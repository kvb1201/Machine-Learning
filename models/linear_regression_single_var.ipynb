```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
print(df)

```

```python
plt.xlabel("year")
plt.ylabel("per capita in USD")

plt.scatter(df[df.columns[0]],df[df.columns[1]],color ='red',marker='+')



```

```python
reg = linear_model.LinearRegression()
reg.fit(df[[df.columns[0]]],df[df.columns[1]])

reg.predict([[2014]])
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
plt.scatter(df[df.columns[0]],df[df.columns[1]],color ='red',marker='+')
plt.plot(df[df.columns[0]], reg.predict(df[[df.columns[0]]]))

```

```python
import pickle
with open ('model_pikel','wb') as f:
    pickle.dump(reg,f)
```

```python
f = open('model_pikel','rb')
model = pickle.load(f)
```

```python
model.predict([[5000]])
```
