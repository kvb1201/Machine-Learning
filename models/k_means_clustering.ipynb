```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../datasets/income.csv')
df.head()
```

```python
plt.scatter(df.Age, df['Income($)'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_pred = km.fit_predict(df.drop('Name', axis =1))
```

```python
y_pred
```

```python
df['cluster'] = y_pred
df.head()
```

```python
df_0 = df[df.cluster == 0]
df_1 = df[df.cluster == 1]
df_2 = df[df.cluster == 2]


plt.scatter(df_0.Age, df_0['Income($)'], color = 'red')
plt.scatter(df_1['Age'], df_1['Income($)'], color = 'green')
# plt.scatter( df_2['Age'], df['Income($)'], color = 'blue')

plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
plt.show()
```
